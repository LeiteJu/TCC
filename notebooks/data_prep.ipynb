```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
PATH="https://raw.githubusercontent.com/LeiteJu/TCC/main/dados/input/"

x = pd.read_csv(f"{PATH}input.csv")
x = x.sort_values(["data", "estados"])

y = pd.read_csv(f"{PATH}target.csv")
y['data'] = pd.to_datetime(y['data'])
```

```python
ANUAL = [ 'pib_pc', 'pib_pmc', 'pib_pcpt', 'pib_cc',
          'populacao', 'idh_l', 'idh_r', 'idh_e', 'desemprego']
MENSAL = ['incc', 'ipca', 'igp', 'selic', 'nfsp',
           'estoque', 'producao', 'preco_ton', 'preco_saco',
           'preco_kg']
```

```python
NULL = -1
```

```python
df = x.copy()
```

```python
df['data'] = pd.to_datetime(df['data'])
```

```python
df = df[df['data'].dt.year >= 2002]
```

```python
df = pd.merge(df, y, how='left', on=['data', 'estados'])
```

```python
df_sorted = df.sort_values(by=["estados", "data"]).copy()
for estado in df.estados.unique():
    indexes = df_sorted[df_sorted["estados"] == estado].index
    df_sorted.loc[indexes, ANUAL] = df_sorted.loc[indexes, ANUAL].shift(12)
    df_sorted.loc[indexes, MENSAL] = df_sorted.loc[indexes, MENSAL].shift(1)
```

```python
df_sorted.head(15)
```

```python
df_sorted = df_sorted.fillna(NULL)
```

```python
df_sorted = df_sorted[df_sorted['data'].dt.year >= 2003]
```

```python
df_sorted.to_csv('../dados/input/processado.csv', index=False)
```
