Nesse notebook vamos aplicar uma técnica de aumentação de dados para lidar com a pouca quantidade de dados disponíveis

Para cada estado, vamos calcular a média e o desvio padrão dos dados nos 30 anos e vamos multiplicar por 10 a quantidade de entradas, gerando números com uma Gaussiana

Utilizamos o método `random.gauss`



```python
import pandas as pd
import random
from statistics import mean, stdev

import warnings
warnings.filterwarnings('ignore')
```

```python
def init ():

  global df, anomes, n_estados, anomes_aug, path

  path= "/content/drive/MyDrive/TCC/dados/anualizados/csv/"

  # variaveis auxiliares
  anos = [str(i) for i in range(1991, 2020)]
  anos_aug = [str(i) for i in range(1720,1991)]

  meses = ["{:02d}".format(i) for i in range(1,13)] # colocamos os 0 a esquerda se precisar

  anomes = []
  anomes_aug = []
  
  n_estados = 27

  for ano in anos:
    anomes += [str(ano) + i for i in meses]

  for ano in anos_aug:
    anomes_aug += [str(ano) + i for i in meses]

  df = [[i] for i in range(1,28)] * len(anomes)

  df_aug = [[i] for i in range(1,28)] * len(anomes)
```

```python
# gera os valores aleatoriamente, segundo a Gaussiana, para cada estado e indicador
# recebe um dataframe com o indicador
def generate_data (ind):

  global n_estados, anomes_aug

  ind_aug = ind.copy()

  states_ind = {}
  for i in range(n_estados):
     states_ind[str(i)] = [mean(list(ind.iloc[i, 1:].astype(float))), stdev(list(ind.iloc[i, 1:].astype(float)))]

  for i in range(len(anomes_aug)):

    ano = []

    for state in range(n_estados):

      ano += [random.gauss(states_ind[str(state)][0], states_ind[str(state)][1])]

    ind_aug[anomes_aug[i]] = ano

  return ind_aug

```

```python
# inicializando os valores

init()
```

```python
# gerando os novos dados

pib_cte = pd.read_csv(path + "pib_precos_constantes.csv")
pib_cte_aug = generate_data(pib_cte)
pib_cte_aug.to_csv("/content/drive/MyDrive/TCC/dados/aumentados/pib_precos_constantes.csv", index=False)

# PIB a precos de mercado corrente 
pib_pm = pd.read_csv(path + "pib_precos_mercado_corrente.csv")
pib_pm_aug = generate_data(pib_pm)
pib_pm_aug.to_csv("/content/drive/MyDrive/TCC/dados/aumentados/pib_precos_mercado_corrente.csv", index=False)

# PIB a precos de mercado corrente 
pib_pc = pd.read_csv(path + "pib_per_capita.csv")
pib_pc_aug = generate_data(pib_pc)
pib_pc_aug.to_csv("/content/drive/MyDrive/TCC/dados/aumentados/pib_per_capita.csv", index=False)

# PIB a precos de mercado corrente 
pib_cc = pd.read_csv(path + "pib_construcao.csv")
pib_cc_aug = generate_data(pib_cc)
pib_cc_aug.to_csv("/content/drive/MyDrive/TCC/dados/aumentados/pib_construcao.csv", index=False)

pop = pd.read_csv(path + "populacao.csv")
pop_aug = generate_data(pop)
pop_aug.to_csv("/content/drive/MyDrive/TCC/dados/aumentados/populacao.csv", index=False)

ipca = pd.read_csv(path + "IPCA.csv")
ipca_aug = generate_data(ipca)
ipca_aug.to_csv("/content/drive/MyDrive/TCC/dados/aumentados/IPCA.csv", index=False)

incc = pd.read_csv(path + "INCC.csv")
incc_aug = generate_data(incc)
incc_aug.to_csv("/content/drive/MyDrive/TCC/dados/aumentados/INCC.csv", index=False)

selic = pd.read_csv(path + "SELIC.csv")
selic_aug = generate_data(selic)
selic_aug.to_csv("/content/drive/MyDrive/TCC/dados/aumentados/SELIC.csv", index=False)

# IDH Renda
idh_renda = pd.read_csv(path + "idh_renda.csv")
idh_renda_aug = generate_data(idh_renda)
idh_renda_aug.to_csv("/content/drive/MyDrive/TCC/dados/aumentados/idh_renda.csv", index=False)


# IDH Saúde
idh_saude = pd.read_csv(path + "idh_saude.csv")
idh_saude_aug = generate_data(idh_saude)
idh_saude_aug.to_csv("/content/drive/MyDrive/TCC/dados/aumentados/idh_saude.csv", index=False)


# IDH Educação
idh_educacao = pd.read_csv(path + "idh_educacao.csv")
idh_educacao_aug = generate_data(idh_educacao)
idh_educacao_aug.to_csv("/content/drive/MyDrive/TCC/dados/aumentados/idh_educacao.csv", index=False)
```
