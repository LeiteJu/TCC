# Implementação de Rede Neural com PyTorch

Seguindo o tutorial, vamos implementar uma rede neural com PyTorch para prever o clima na Austrália, com fins didáticos

Mais específicamente, implementar, treinar e utilizar uma Rede Neural Feed-Foward simples para uma tarefa de classificação binária

```python
# importando as libs
import pandas as pd
import numpy as np
import sklearn
import torch
import seaborn as sns
```

```python
# importando os dados
raw_data=pd.read_csv("weatherAUS.csv")
```

```python
# inspecting the data
raw_data.head()
```

```python
raw_data.describe()
```

Escolhendo as variáveis meteorológicas que interessam: **Rainfall, Humidity3pm, Pressure9am e RainToday**, essas são nossas *features*




```python
cols = ['Rainfall', 'Humidity3pm', 'Pressure9am', 'RainToday', 'RainTomorrow']
processed_data = raw_data[cols]
```

## Pré-processamnto

- *RainToday* e *RainTomorrow*: tem valores "yes" ou "no" vamos converter para 1 e 0 e dropar os valores que não tem entradas

```python
# o output aqui se trata de um aviso, não um erro, basta ignorar por hora...

# substituindo 'yes' e 'no' po 1 e 0
processed_data['RainToday'].replace({'No':0, 'Yes':1}, inplace=True)
processed_data['RainTomorrow'].replace({'No':0, 'Yes':1}, inplace=True)
```

```python
# dropando entradas sem valores
processed_data = processed_data.dropna(how='any')
```
